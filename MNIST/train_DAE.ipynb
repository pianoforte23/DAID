```python
import numpy as np
import os
import tensorflow as tf
import keras
from keras.callbacks import EarlyStopping, ModelCheckpoint
import setup_mnist
```

### Data preprocessing (MNIST)

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
image_size = x_train.shape[1]
x_train, x_test = x_train / 255.0, x_test / 255.0
x_train = np.reshape(x_train, [-1, image_size, image_size, 1])
x_test = np.reshape(x_test, [-1, image_size, image_size, 1])
```

### Denoising autoencoder (DAE) training

```python
input_shape = (28, 28, 1)

DAE = setup_mnist.mnist_dae(input_shape=input_shape)
DAE_path = 'model/DAE.h5'
DAE.compile(loss='mse', optimizer='adam')

nb_epochs = 200
batch_size = 256
```

```python
def train_DAE(DAE, DAE_path):
    if not os.path.exists(DAE_path):
        train_noise = np.random.normal(loc=0, scale=0.1, size=x_train.shape)
        test_noise = np.random.normal(loc=0, scale=0.1, size=x_test.shape)
        x_train_noise = np.clip(x_train + train_noise, 0, 1)
        x_test_noise = np.clip(x_test + test_noise, 0, 1)
        callbacks = [EarlyStopping(monitor='val_loss', patience=20), ModelCheckpoint(DAE_path, monitor='val_loss', save_best_only=True)]
        DAE.fit(x_train_noise, x_train, batch_size=batch_size, validation_data=(x_test_noise, x_test), epochs=nb_epochs, callbacks=callbacks)
```

```python
train_DAE(DAE, DAE_path)
```
