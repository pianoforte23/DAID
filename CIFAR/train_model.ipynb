```python
import numpy as np
import os
import keras
import setup_cifar
from keras.datasets import cifar10
from keras.preprocessing.image import ImageDataGenerator
from keras.callbacks import EarlyStopping, ModelCheckpoint
```

### Data preprocessing (CIFAR-10)

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
x_train = x_train.astype('float32') / 255
x_test = x_test.astype('float32') / 255
```

```python
datagen = ImageDataGenerator(
    featurewise_center=False,
    samplewise_center=False,
    featurewise_std_normalization=False,
    samplewise_std_normalization=False,
    zca_whitening=False,
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    vertical_flip=False
)
datagen.fit(x_train)
```

### Hyperparameter setting & CIFAR-10 model training

```python
loss_object = keras.losses.SparseCategoricalCrossentropy(from_logits=True)
input_shape = (32, 32, 3)

model_resnet = setup_cifar.resnet_v2(input_shape, 20)
model_resnet.compile(loss=loss_object, optimizer='adam', metrics=['accuracy'])

model_vgg16 = setup_cifar.model_vgg16(input_shape)
model_vgg16.compile(loss=loss_object, optimizer='adam', metrics=['accuracy'])

model_resnet_path = "model/model_resnet.h5"
model_vgg16_path = "model/model_vgg16.h5"

callbacks_resnet = [EarlyStopping(monitor='val_accuracy', patience=50), ModelCheckpoint(model_resnet_path, monitor='val_accuracy', save_best_only=True)]
callbacks_vgg16 = [EarlyStopping(monitor='val_accuracy', patience=50), ModelCheckpoint(model_vgg16_path, monitor='val_accuracy', save_best_only=True)]

nb_epochs = 200
batch_size = 256
```

```python
if not os.path.exists(model_resnet_path):
    model_resnet.fit_generator(datagen.flow(x_train, y_train, batch_size=batch_size), steps_per_epoch=x_train.shape[0] // batch_size, validation_data=(x_test, y_test), epochs=nb_epochs, callbacks=callbacks_resnet)
```

```python
if not os.path.exists(model_vgg16_path):
    model_vgg16.fit_generator(datagen.flow(x_train, y_train, batch_size=batch_size), steps_per_epoch=x_train.shape[0] // batch_size, validation_data=(x_test, y_test), epochs=nb_epochs, callbacks=callbacks_vgg16)
```
